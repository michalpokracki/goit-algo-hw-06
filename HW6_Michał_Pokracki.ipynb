# Task 1: Create and Visualize a Graph

```python
import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Create the graph
G = nx.Graph()

# Add nodes (stations)
stations = ['A', 'B', 'C', 'D', 'E', 'F']
G.add_nodes_from(stations)

# Add edges (connections)
connections = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('C', 'E'), ('D', 'F'), ('E', 'F')]
G.add_edges_from(connections)

# Step 2: Visualize the graph
plt.figure(figsize=(10, 7))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=2000, font_size=20, font_color='black', font_weight='bold', edge_color='gray')
plt.title("City Transportation Network")
plt.show()

# Step 3: Analyze the graph
num_vertices = G.number_of_nodes()
num_edges = G.number_of_edges()
degrees = dict(G.degree())

print(f"Number of vertices: {num_vertices}")
print(f"Number of edges: {num_edges}")
print("Degrees of each vertex:", degrees)
```

# Task 2: Implement DFS and BFS

```python
def dfs_paths(graph, start, goal):
    stack = [(start, [start])]
    while stack:
        (vertex, path) = stack.pop()
        for next in set(graph[vertex]) - set(path):
            if next == goal:
                yield path + [next]
            else:
                stack.append((next, path + [next]))

def bfs_paths(graph, start, goal):
    queue = [(start, [start])]
    while queue:
        (vertex, path) = queue.pop(0)
        for next in set(graph[vertex]) - set(path):
            if next == goal:
                yield path + [next]
            else:
                queue.append((next, path + [next]))

# Example of usage
start_node, end_node = 'A', 'F'
dfs_result = list(dfs_paths(G, start_node, end_node))
bfs_result = list(bfs_paths(G, start_node, end_node))

print("DFS paths from A to F:", dfs_result)
print("BFS paths from A to F:", bfs_result)
```

# Task 3: Implement Dijkstra's Algorithm

```python
# Step 1: Add weights to the edges
weighted_connections = [
    ('A', 'B', 1), ('A', 'C', 5),
    ('B', 'D', 3), ('C', 'D', 2),
    ('C', 'E', 2), ('D', 'F', 1),
    ('E', 'F', 4)
]
G_weighted = nx.Graph()
G_weighted.add_weighted_edges_from(weighted_connections)

# Step 2: Implement Dijkstra's Algorithm
def dijkstra_path(graph, start, goal):
    return nx.dijkstra_path(graph, start, goal)

# Example of usage
shortest_path = dijkstra_path(G_weighted, start_node, end_node)
print("Shortest path from A to F using Dijkstra's algorithm:", shortest_path)
```
